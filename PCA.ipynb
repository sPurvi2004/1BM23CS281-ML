```python
# Manual PCA with features (matches notebook exactly)

import numpy as np

# -------------------------------
# Step 1: Features (same as your page)
# -------------------------------
feature1 = [1, 2, 3, 4, 5]
feature2 = [5, 4, 3, 2, 1]
feature3 = [10, 8, 6, 4, 2]

print("features 1 :", feature1)
print("features 2 :", feature2)
print("features 3 :", feature3)

# Convert to matrix
X = np.array([
    feature1,
    feature2,
    feature3
]).T   # transpose to match rows

# -------------------------------
# Step 2: Mean center
# -------------------------------
mean = np.mean(X, axis=0)
X_centered = X - mean

# -------------------------------
# Step 3: Covariance
# -------------------------------
cov_matrix = np.cov(X_centered, rowvar=False)

# -------------------------------
# Step 4: Eigenvalues & vectors
# -------------------------------
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort
idx = np.argsort(eigenvalues)[::-1]
eigenvectors = eigenvectors[:, idx]

# -------------------------------
# Step 5: Select top 2 components
# -------------------------------
W = eigenvectors[:, :2]

# -------------------------------
# Step 6: Transform
# -------------------------------
reduced_data = X_centered @ W

# -------------------------------
# Step 7: Output EXACT format
# -------------------------------
print("\nReduced Data\n")
for row in reduced_data:
    print("[", round(row[0], 2), " ", round(row[1], 2), "]")
```
