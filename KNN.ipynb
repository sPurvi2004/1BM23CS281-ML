```python
# KNN Classification (Exact notebook format)

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

print("Output for KNN Classification\n")

# Heading exactly like your page
print("Sepal        Sepal        petal        petal")
print("length       width        length       width")

# Print first 5 rows
for i in range(5):
    print(i, "   ", X[i][0], "   ", X[i][1], "   ", X[i][2], "   ", X[i][3])

# Target
print("\nTarget\n")
for i in range(5):
    print(i, "   ", y[i])

# Model
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

# New flower
new_flower = [[5, 3.5, 1.3, 0.2]]

# Prediction
pred = model.predict(new_flower)

# Final output EXACT wording
print("\nNew flower data:", new_flower)
print("Predicted class index:", pred[0])
print("Predicted class name:", iris.target_names[pred][0])
```

```python
# KNN Regression (Exact output format)

from sklearn.datasets import load_diabetes
from sklearn.neighbors import KNeighborsRegressor

# Load dataset
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

# Model (k = 3)
model = KNeighborsRegressor(n_neighbors=3)

# Train model
model.fit(X, y)

# New patient data (same as your notebook)
new_patient = [[-0.001882, -0.044642, -0.051474, -0.026328,
                -0.008449, -0.019163, 0.074412, -0.039493,
                -0.06833, -0.092204]]

# Predict
prediction = model.predict(new_patient)

# Output EXACT format
print("Output for KNN Regressor\n")
print("New patient data:", new_patient)
print("\nPredicted disease progression :", round(prediction[0], 2))
```
